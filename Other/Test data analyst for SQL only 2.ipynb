```python
# загрузка нужных библиотек
from io import BytesIO
import requests
import pandas as pd
import pandasql as ps

# убирает ограничение по демонстрации колонок
pd.set_option('display.max_columns', None)
```

```python
# создание функции для загрузки таблиц с Google Sheets

def google(spreadsheet_id, sheet_name):

    file_name = (
        'https://docs.google.com/spreadsheets/d/{}/export?format=xlsx'
        .format(spreadsheet_id)
        )
    r = requests.get(file_name)
    df = pd.read_excel(BytesIO(r.content), sheet_name=sheet_name)

    return df
```

```python
# чтение данных и запись в переменные
report_offer = google('1jTXLH6jNiPOgm6tArgdGU8oQ86rgHUe-HkVTNp8JiUQ', 'ReportOffer')
buyer = google('1jTXLH6jNiPOgm6tArgdGU8oQ86rgHUe-HkVTNp8JiUQ', 'Buyer')
offer = google('1jTXLH6jNiPOgm6tArgdGU8oQ86rgHUe-HkVTNp8JiUQ', 'Offer')
offer_conv_target = google('1jTXLH6jNiPOgm6tArgdGU8oQ86rgHUe-HkVTNp8JiUQ', 'OfferConversionTarget')
conv_target = google('1jTXLH6jNiPOgm6tArgdGU8oQ86rgHUe-HkVTNp8JiUQ', 'ConversionTarget')
```

```python
# создание функции для первичного обзора данных

def first_check(data):

    print('Первые 10 строк таблицы \n')
    display(data.head(10))

    print('-' * 50)
    print('Последние 10 строк таблицы \n')
    display(data.tail(10))

    print('-' * 50)
    print('Информация о таблице \n')
    print(data.info())

    print('-' * 50)
    print('Количество дубликатов:', data.duplicated().sum())
```

**FYI (disclaimer)**

1. Для компенсации невозможности объединения таблиц ReportOffer и OfferConversionTarget по полю traffic_source, для каждого offer_id была отобрана кампания с наибольшей длительностью, чтобы избежать дубликатов, поскольку длительность кампании как раз отличается источником трафика в поле traffic_source.

2. По аналогичной причине не использовалось поле cost из OfferConversionTarget, поскольку оно завязано на поле traffic_source. Если его использовать, то образуются ненужные дубликаты. Если бы его можно было использовать, то был бы осуществлён перевод валюты в рубли и данная сумма использовалась бы в качестве затрат.

3. В поле platform таблицы OfferConversionTarget есть значение 'Undefined' аналогичное тому, что есть в поле country_list. Поскольку в задании говорилось только про то, что страна Undefined связывается с любой страной, к platform такое условие не применялась и замена значения 'Undefined' не производилась.

4. Поскольку нет описания того, какие поля являются доходом, а какие - расходом, то поле traffic_cost из ReportOffer использовалось как расходы, а поле payout - как доходы.

```python
query = '''
    WITH oft AS ( -- сбор информации об офферах
                SELECT *,
                        -- определение продолжительности кампании для удаления дубликатов
                        julianday(date_to) - julianday(date_since) AS diff

                FROM ( SELECT DISTINCT
                              o.id,
                              o.title,
                              o.tracker,
                              o.external_id,
                              date(ot.date_since) AS date_since,

                              -- заполнение пустот date_to текущей датой
                              CASE
                                  WHEN date_to IS NULL THEN date('now')
                                  ELSE date(ot.date_to)
                              END AS date_to,

                              ot.platform,
                              ot.country_list,

                              -- замена страны 'Undefined' на список стран, которые упоминаются в report_offer
                              CASE
                                  WHEN ot.country_list = 'Undefined' THEN (SELECT GROUP_CONCAT(DISTINCT country) FROM report_offer)
                                  ELSE ot.country_list
                              END AS country_raw

                      FROM offer AS o

                      INNER JOIN (SELECT DISTINCT * FROM offer_conv_target) AS ot
                      ON o.id = ot.offer_id

                      INNER JOIN (SELECT DISTINCT * FROM conv_target) AS ct
                      ON ot.conversion_target_id = ct.id
                      AND ct.code IN ('installs', 'is_traffic')

                      WHERE o.tracker = 'Appsflyer'
                      ) AS oft_raw
                ),

      ctr AS (  -- таблица с максимальной продолжительностью кампаний и списком стран
                -- для каждого offer_id и соответствующей платформы
                SELECT  id,
                        platform,
                        MAX(diff) AS max_diff,
                        GROUP_CONCAT(DISTINCT country_raw) AS country
                  FROM oft
                  GROUP BY 1, 2
                      ),

      o_fin AS ( -- получение таблицы с нужными офферами
                SELECT  DISTINCT
                        oft.id,
                        oft.title,
                        oft.tracker,
                        oft.external_id,
                        oft.date_since,
                        oft.date_to,
                        oft.platform,
                        ctr.country

                FROM oft

                -- удаление лишних кампаний через объединение таблиц
                INNER JOIN ctr
                ON oft.id = ctr.id
                AND oft.platform = ctr.platform
                AND oft.diff = ctr.max_diff
                ),

        b AS (-- убирает дубликаты в buyer через дату последних изменений
                SELECT  DISTINCT
                        id,
                        FIRST_VALUE(buyer_name) OVER(PARTITION BY id ORDER BY updated_at DESC) AS buyer_name

                FROM (SELECT  id,

                              CASE
                                  WHEN name = 'Default Buyer' OR  hasoffers_id = 'DefaultBuyer' THEN 'Ada Lovelace [130]'
                                  ELSE name || ' [' || hasoffers_id || ']'
                              END AS buyer_name,

                              hasoffers_id,
                              updated_at

                      FROM buyer
                ) AS a_raw
              )

        -- объединение всех таблиц и расчёт числовых данных
        SELECT  r.date,
                b.buyer_name,
                o_fin.id AS offer_title,
                o_fin.tracker AS tracker_type,
                o_fin.external_id AS external_offer_id,
                r.country,
                r.platform,
                r.traffic_source,
                SUM(r.payout) AS revenue,
                SUM(r.traffic_cost) AS costs,
                SUM(r.payout - r.traffic_cost) AS margin,
                SUM(r.confirmed_conversion) AS conversion_cnt

        FROM report_offer AS r

        -- добавляет информацию о байерах
        INNER JOIN b
        ON r.buyer_id = b.id

        -- добавляет информацию об офферах
        INNER JOIN o_fin
        ON r.offer_id = o_fin.id
        AND r.date BETWEEN o_fin.date_since AND o_fin.date_to
        AND o_fin.country LIKE '%' || r.country || '%'
        AND r.platform = o_fin.platform

        WHERE DATE(r.date, 'start of month') = '2024-04-01' -- данные за текущий месяц

        GROUP BY 1, 2, 3, 4, 5, 6, 7, 8

        -- убирает строки с нулевой маржой
        HAVING SUM(r.payout - r.traffic_cost) <> 0

    '''
query_res = ps.sqldf(query, locals())

query_res
```
